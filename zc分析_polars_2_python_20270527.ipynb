# 增加冷门2

```python

import polars as pl
low_value = 3.1
high_value = 4.0
big_value = 5.0

# 假设 df 是已经存在的 Polars DataFrame
df = df.with_columns(pl.lit("正常").alias("冷门2"))
# 根据条件更新 '冷门2' 列的值
df = df.with_columns(
    pl.when(
        (pl.col("终负赔") >= pl.col("终胜赔")) &
        (~pl.col("终盘让球").str.contains("受")) &
        (pl.col("终负赔") > low_value) &
        (pl.col("终负赔") <= high_value) &
        (pl.col("结果") != 3)
    ).then(pl.lit("主小冷"))
    .when(
        (pl.col("终负赔") >= pl.col("终胜赔")) &
        (~pl.col("终盘让球").str.contains("受")) &
        (pl.col("终负赔") > high_value) &
        (pl.col("终负赔") <= big_value) &
        (pl.col("结果") != 3)
    ).then(pl.lit("主中冷"))
    .when(
        (pl.col("终负赔") >= pl.col("终胜赔")) &
        (~pl.col("终盘让球").str.contains("受")) &
        (pl.col("终负赔") > big_value) &
        (pl.col("结果") != 3)
    ).then(pl.lit("主大冷"))
#     .otherwise(pl.col("冷门2"))
#     .alias("冷门2")
    .when(
        (pl.col("终负赔") <= pl.col("终胜赔")) &
        (pl.col("终盘让球").str.contains("受")) &
        (pl.col("终胜赔") > low_value) &
        (pl.col("终胜赔") <= high_value) &
        (pl.col("结果") != 0)
    ).then(pl.lit("客小冷"))
    .when(
        (pl.col("终负赔") <= pl.col("终胜赔")) &
        (pl.col("终盘让球").str.contains("受")) &
        (pl.col("终胜赔") > high_value) &
        (pl.col("终胜赔") <= big_value) &
        (pl.col("结果") != 0)
    ).then(pl.lit("客中冷"))
    .when(
        (pl.col("终负赔") <= pl.col("终胜赔")) &
        (pl.col("终盘让球").str.contains("受")) &
        (pl.col("终胜赔") > big_value) &
        (pl.col("结果") != 0)
    ).then(pl.lit("客大冷"))
    .otherwise(pl.col("冷门2"))
    .alias("冷门2")    
)

# 对于 '主队进球' == -1 的情况，将 '冷门' 列的值更新为 '正常'
df = df.with_columns(
    pl.when(pl.col("主队进球") == -1)
    .then(pl.lit("正常"))
    .otherwise(pl.col("冷门2"))
    .alias("冷门2")
)
df.tail(15)
# Now df contains the updated column 冷门
```

```python
import polars as pl
import numpy as np
import itertools

ranges = {
    "终盘主水": (0.75, 1.2),
    "终盘水差": (0.05, 0.4),
    "主队净胜球": (-2, 4),
}

def generate_range_combinations(ranges, step):
    combinations = []
    for col, (min_val, max_val) in ranges.items():
        col_ranges = np.arange(min_val, max_val + step, step)
        combinations.append([(col, start, start + step) for start in col_ranges])
    return combinations

best_max_count = 0
best_combination = None
best_step = None

for step in [0.05, 0.1, 0.15, 0.2]:  # 尝试不同的 step 值
    combinations = generate_range_combinations(ranges, step)
    temp_all = []

    for comb in itertools.product(*combinations):
        temp_df = df
        temp_count = []
        filters = None

        for col, start, end in comb:
            filter_condition = (pl.col(col) >= start) & (pl.col(col) < end)
            if filters is None:
                filters = filter_condition
            else:
                filters = filters & filter_condition

        temp_df = temp_df.filter(filters)
        temp_count.append([temp_df.filter(pl.col("冷门").str.contains("冷"))["冷门"].count(), start, end,step])

        count = temp_df.filter(pl.col("冷门").str.contains("冷")).count()["冷门"]
        temp_all.append(temp_count)

        if count.max() > best_max_count:
            best_max_count = count.max()
            best_combination = comb
            best_step = step

sorted_temp_all = sorted(temp_all, key=lambda x: x[0][0], reverse=True)[:5]
print(sorted_temp_all)
print(f"最佳组合: {best_combination}, 包含'冷'的'冷门'总数: {format(best_max_count, '.2f')}, 最佳 step 值: {best_step}")
```
