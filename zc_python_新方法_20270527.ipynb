```python
# 定义函数来检查值并返回不同的结果
def check_values(row):
    if row['欧赔判断'] == row['亚赔判断']:
        all_values = {3, 1, 0}
        return list(all_values - {row['欧赔判断']})
    else:
        all_values = {3, 1, 0}
        return list(all_values - {row['欧赔判断'], row['亚赔判断']})

# 应用函数
df['自判'] = df.apply(check_values, axis=1)
```
