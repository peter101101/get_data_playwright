```python
import polars as pl
low_value = 3.1
high_value = 4.0
big_value = 5.0

# 假设 df 是已经存在的 Polars DataFrame
df=dfl
df = df.with_columns(pl.lit("正常").alias("冷门2"))

# 根据条件更新 '冷门2' 列的值
df = df.with_columns(
    pl.when(
        (pl.col("终负赔") >= pl.col("终胜赔")) &
        (~pl.col("终盘让球").str.contains("受")) &
        (pl.col("终负赔")-(pl.col("终负赔")) <= 0.3) &
        (pl.col("终负赔") <= high_value) &
        (pl.col("结果") != 3)
    ).then(pl.lit("主小冷"))
    .when(
        (pl.col("终负赔") >= pl.col("终胜赔")) &
        (~pl.col("终盘让球").str.contains("受")) &
        (pl.col("终负赔") > high_value) &
        (pl.col("终负赔") <= big_value) &
        (pl.col("结果") != 3)
    ).then(pl.lit("主中冷"))
    .when(
        (pl.col("终负赔") >= pl.col("终胜赔")) &
        (~pl.col("终盘让球").str.contains("受")) &
        (pl.col("终负赔") > big_value) &
        (pl.col("结果") != 3)
    ).then(pl.lit("主大冷"))
#     .otherwise(pl.col("冷门2"))
#     .alias("冷门2")
    .when(
        (pl.col("终负赔") <= pl.col("终胜赔")) &
        (pl.col("终盘让球").str.contains("受")) &
        (pl.col("终胜赔") > low_value) &
        (pl.col("终胜赔") <= high_value) &
        (pl.col("结果") != 0)
    ).then(pl.lit("客小冷"))
    .when(
        (pl.col("终负赔") <= pl.col("终胜赔")) &
        (pl.col("终盘让球").str.contains("受")) &
        (pl.col("终胜赔") > high_value) &
        (pl.col("终胜赔") <= big_value) &
        (pl.col("结果") != 0)
    ).then(pl.lit("客中冷"))
    .when(
        (pl.col("终负赔") <= pl.col("终胜赔")) &
        (pl.col("终盘让球").str.contains("受")) &
        (pl.col("终胜赔") > big_value) &
        (pl.col("结果") != 0)
    ).then(pl.lit("客大冷"))
    .otherwise(pl.col("冷门2"))
    .alias("冷门2")    
    )


# 对于 '主队进球' == -1 的情况，将 '冷门' 列的值更新为 '正常'
df = df.with_columns(
    pl.when(pl.col("主队进球") == -1)
    .then(pl.lit("正常"))
    .otherwise(pl.col("冷门2"))
    .alias("冷门2")
    )

df = df.with_columns(
    (pl.col("终胜赔")-pl.col("终平赔")).alias("胜平差"),
    (pl.col("终平赔")-pl.col("终负赔")).alias("平负差"),
    (pl.col("终胜赔")-pl.col("终负赔")).alias("胜负差")
    )
    
    

df.tail(5)
```

```python
def calculate_result(df):
    # 定义新列"结果"的计算逻辑
    conditions = [
    # 规则1
    (
    (df["胜平差"] < 0) &
    (df["平负差"]  < 0) &
    (df["胜负差"]  < 0) &
    # (df["终胜赔"] - df["终平赔"] >= -1.3) &
    # (df["终胜赔"] - df["终平赔"] <= -1.9) ,
    (df["胜负差"] >= -1.9) &
    (df["胜负差"] <= -1.35) &
    (df["胜平差"] >= -1.99) &
    (df["胜平差"] <= -0.42), 
#     (df["平负差"] >= -3.08) &
#     (df["平负差"] <= -1.8) ,        
    pl.lit(0)
        ),
    ]
    # 规则2
    # (
    # (df["终胜赔"] - df["终平赔"] < 0) &
    # (df["终平赔"] - df["终负赔"] > 0) &
    # (df["终胜赔"] - df["终负赔"] < 0) &
    # (df["终胜赔"] - df["终平赔"] < -1.29) &
    # (df["终胜赔"] - df["终负赔"] > 0.45) &
    # (df["终平赔"] - df["终负赔"] <= -0.84),
    # pl.lit(0)
    # ),
    # 更多规则...
    # 规则7
    # (
    # (df["终胜赔"] - df["终平赔"] <= 0) &
    # (df["终平赔"] - df["终负赔"] <= 0) &
    # (df["终胜赔"] - df["终负赔"] <= 0) &
    # (df["终胜赔"] - df["终平赔"] < -0.5) &
    # (df["终胜赔"] - df["终负赔"] >= 1.37) &
    # (df["终平赔"] - df["终负赔"] >= 0.87),
    # pl.lit(3)
    # ),
    # # 规则9
    # (
    # (df["终胜赔"] - df["终平赔"] < -0.38) &
    # (df["终胜赔"] - df["终负赔"] > 2.76) &
    # (df["终平赔"] - df["终负赔"] >= 2.38),
    # pl.lit(1)
    # ),
    # # 规则10
    # (
    # (df["终胜赔"] - df["终平赔"] <= 0) &
    # (df["终平赔"] - df["终负赔"] <= 0) &
    # (df["终胜赔"] - df["终负赔"] <= 0) &
    # (df["终胜赔"] - df["终平赔"] < 0.3) &
    # (df["终胜赔"] - df["终负赔"] > 1.62) &
    # (df["终平赔"] - df["终负赔"] >= 1.92),
    # pl.lit(1)
    # ),
    # 默认规则
#     (
#     pl.lit(True),
#     pl.lit(None)
#     )
#     ]

    # 应用规则
    result_expr = pl.when(conditions[0][0]).then(conditions[0][1])
    for condition, value in conditions[1:]:
        result_expr = result_expr.when(condition).then(value)
        result_expr = result_expr.otherwise(pl.lit(None))

    # 添加"结果"列
    df = df.with_columns(result_expr.alias("结果2"))

    return df

# 假设df是您的DataFrame
# df = pl.DataFrame({...})

# 计算结果
df_result = calculate_result(df)
df_result.select(pl.col(["结果","结果2"]))
```

```python
df_gd_主冷 = (
    df.group_by(['结果', '冷门',"终水位分类"])
    .agg(
        [
            pl.col('胜价值').max().alias("胜价值_max"),
            pl.col('胜价值').min().alias("胜价值_min"),
            pl.col('胜价值').mean().alias("胜价值_mean"),
            pl.col('平价值').max().alias("平价值_max"),
            pl.col('平价值').min().alias("平价值_min"),
            pl.col('平价值').mean().alias("平价值_mean"),
            pl.col('负价值').max().alias("负价值_max"),
            pl.col('负价值').min().alias("负价值_min"),
            pl.col('负价值').mean().alias("负价值_mean"),
            pl.col('冷门').count().alias("冷门_count"),
        ]
    ).filter(pl.col("冷门").str.contains("^主")).sort("冷门",descending=True)
)

df_gd_客冷 = (
    df.group_by(['结果', '冷门',"终水位分类"])
    .agg(
        [
            pl.col('胜价值').max().alias("胜价值_max"),
            pl.col('胜价值').min().alias("胜价值_min"),
            pl.col('胜价值').mean().alias("胜价值_mean"),
            pl.col('平价值').max().alias("平价值_max"),
            pl.col('平价值').min().alias("平价值_min"),
            pl.col('平价值').mean().alias("平价值_mean"),
            pl.col('负价值').max().alias("负价值_max"),
            pl.col('负价值').min().alias("负价值_min"),
            pl.col('负价值').mean().alias("负价值_mean"),
            pl.col('冷门').count().alias("冷门_count"),
        ]
    ).filter(pl.col("冷门").str.contains("^客")).sort("冷门",descending=True)
)

df_gd_正常 = (
    df.group_by(['结果', '冷门',"终水位分类"])
    .agg(
        [
            pl.col('胜价值').max().alias("胜价值_max"),
            pl.col('胜价值').min().alias("胜价值_min"),
            pl.col('胜价值').mean().alias("胜价值_mean"),
            pl.col('平价值').max().alias("平价值_max"),
            pl.col('平价值').min().alias("平价值_min"),
            pl.col('平价值').mean().alias("平价值_mean"),
            pl.col('负价值').max().alias("负价值_max"),
            pl.col('负价值').min().alias("负价值_min"),
            pl.col('负价值').mean().alias("负价值_mean"),
            pl.col('冷门').count().alias("冷门_count"),
        ]
    ).filter(pl.col("冷门").str.contains("正常")).sort("冷门",descending=True)
)
df_gd_主冷.to_pandas().to_excel("output_data\df_价值_主冷门_结果.xlsx",
                sheet_name='Sheet1',
                index=True,
                float_format="%.2f")
df_gd_客冷.to_pandas().to_excel("output_data\df_价值_客冷门_结果.xlsx",
                sheet_name='Sheet1',
                index=True,
                float_format="%.2f")
df_gd_正常.to_pandas().to_excel("output_data\df_价值_正常_结果.xlsx",
                sheet_name='Sheet1',
                index=True,
                float_format="%.2f")
```
